7. Convert all categorical features with 2 unique string values to binary features.

>>'doornumber' column name should be converted to 'is_four_door' and the values to 1 or 0.

```

cars_df['is_four_door'] = (cars_df['doornumber'] == 'four').astype(int)
cars_df['is_turbo'] = (cars_df['aspiration'] == 'turbo').astype(int)
cars_df['is_gas'] = (cars_df['fueltype'] == 'gas').astype(int)
cars_df['is_front_engine'] = (cars_df['enginelocation'] == 'front').astype(int)
cars_df.drop(['doornumber', 'aspiration', 'fueltype', 'enginelocation'], axis=1, inplace=True)
```

14. Set the X and y and divide the DF to train and test dataframes

Tip - I used **random_state=1234** if you use a different number you would get different results...

```
from sklearn.model_selection import train_test_split
X= cars_df[correlated]
y= cars_df.priceK

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
train_df = pd.merge(left=X_train, right=y_train, left_index=True, right_index=True)
test_df = pd.merge(left=X_test, right=y_test, left_index=True, right_index=True)

test_df.head(10)

```

18. Create a linear model and train it. Display the model coeffiecients and formula. You can create all types of models (SLR, MLR and polynomial) but choose to use only one. Write down a short explanation why you decided to select the secific model type.

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

mlr_model = LinearRegression()
mlr_model.fit(X_train_scaled, y_train_scaled)

# Display the coefficients
coefficients = mlr_model.coef_
intercept = mlr_model.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)

formula = "y = "
for i in range(len(coefficients)):
    formula += f"({coefficients[i][0]:.2f} * X{str(i+1)}) + "
formula += f"{intercept[0]:.2f}"
print("Formula:", formula)
```

19. Use the model to predict the train prices. Inverse the scale and add the results to the train DF.

```
train_predictions_scaled = mlr_model.predict(X_train_scaled)

train_predictions = y_train_scaler.inverse_transform(train_predictions_scaled)

train_df['predicted_price'] = train_predictions

train_df.head(10)
```
